```python
def get_data(path, columns):
    db     = pd.ExcelFile(path)
    sheets = dict() #DICIONÁRIO COM AS FOLHAS

    for freq in db.sheet_names:
        data = db.parse(freq)[columns]
        sheets[freq] = data
    
    return sheets
```
